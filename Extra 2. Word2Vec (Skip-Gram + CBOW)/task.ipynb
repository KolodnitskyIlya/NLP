# Реализация Word2Vec (Skip Gram / CBoW)

## Задание

* Реализовать алгоритм получения контекстных эмбеддингов (векторных представлений) слов.
* Реализовать алгоритмы CBoW и SkipGram для получения собственных моделей Word2Vec для выбранных данных.
* Визуализировать векторные представления в малой размерности.

## Ход работы

1. Загрузить коллекцию текстовых документов: [просьбы жильцов](https://drive.google.com/file/d/1oU4WZnxmOwp2isH6Sv9yAk2C_48EJQhh/view?usp=sharing). Можете использовать не весь датасет (подмножество строк таблицы). В этом задании вас интересует столбец с текстом обращений жильцов. Можете загрузить любой датасет для классификации текстовых данных.
2. Рассмотреть каждый текстовый документ отдельно. Выполнить предварительную обработку текстовых данных.
3. С помощью PyTorch реализовать алгоритмы построения Word2Vec: CBoW и SkipGram. Получить векторные представления текстов с помощью собственной реализации данных алгоритмов.
4. Сформировать матрицы контекстных эмбеддингов. С помощью PCA уменьшить размер данных матриц (размер задать самостоятельно).
5. Сохранить полученные различными способами векторные представления в файлы tsv. Визуализировать их с помощью сервиса [Projector](https://projector.tensorflow.org/) от авторов TensorFlow
> Вместо Projector можно использовать [TensorBoard](https://www.tensorflow.org/tensorboard/tensorboard_projector_plugin?hl=ru)

## Критерии приема работы

* Загружен / выбран датасет для классификации.
* Представлен код предварительной обработки данных.
* Представлена реализация CBoW или SkipGram.
* Векторные представления сохранены и визуализированы.
