# Тематическое моделирование новостных статей

## Задание

* Познакомиться с инструментом BigARTM для тематического моделирования.
* Применить на практике теоретические знания о тематическом моделировании.
* Построить тематическую модель коллекции новостных статей с применением аддитивной регуляризации.

## Ход работы

> Обратите внимание! Лабораторная работа должна быть сделана при помощи библиотеки **BigARTM** (классический и мощный инструмент тематического моделирования, который можно тонко настраивать). Однако на последних версиях Python библиотека не рабоатет. По этой причине в таких средах, как Google Colab и Kaggle, у вас работать **не получится**... По этой причине лабу необходимо делать в локальных окружениях. Гайд, как установить библиотеку к себе: https://habr.com/ru/sandbox/182930/

1. Загрузите [обучающий](https://huggingface.co/datasets/mlsum/resolve/refs%2Fconvert%2Fparquet/ru/train/0000.parquet?download=true) и [тестовый](https://huggingface.co/datasets/mlsum/resolve/refs%2Fconvert%2Fparquet/ru/test/0000.parquet?download=true) набор данных (две коллекции) новостных статей (с сайта mk.ru).
2. Данные представлены в формате [Apache Parquet](https://habr.com/ru/companies/wrike/articles/279797/), загрузите их в DataFrame с использованием pd.read_parquet. Пример: pd.read_parquet('0000.parquet', 'pyarrow').
3. Выполните предварительную обработку обучающей и тестирующей выборки, преобразуйте обе коллекции документов. Выполните векторизацию с использованием CountVectorizer и получите документы в виде “мешков слов”.
4. Обучите модель LDA с использованием библиотеки BigARTM. С помощью matplotlib постройте график изменения перплексии в ходе обучения.
5. Попробуйте несколько вариантов значений параметра “количество тем” (минимум 3 варианта). Не меняя других параметров модели подберите оптимальное количество тем (с помощью перплексии).
6. Определите темы документов тестовой выборки и расчитайте перплексию для нее.
7. Обучите модель общего вида, добавьте несколько регуляризаторов (минимум один). Постройте график изменения перплексии в ходе обучения.
8. Определите темы документов тестовой выборки и расчитайте перплексию для нее.

## Критерии приема работы

* Данные загружены и предобработаны.
* Выполнена векторизация документов и произведена загрузка коллекции в виде батчей.
* Обучена модель LDA, определены темы для тестовой выборки, есть график перплексии, рассчитана перплексия тестовой выборки.
* Показано использование перплексии для подбора оптимального количества тем.
* Обучена модель общего вида (минимум один регуляризатор), определены темы для тестовой выборки, есть график перплексии, расчитана перплексия тестовой выборки.
