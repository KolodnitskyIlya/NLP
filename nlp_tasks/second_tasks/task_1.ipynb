# Парсинг данных с hh.ru

## Задание

* Самое время положить серваки hh.ru 😈;
* изучить процесс сбора данных для задач обработки естественного языка с использованием многопоточности и асинхронности.

## Ход работы


1. Самостоятельно изучите структуру бесплатного API hh.ru (если требуется, получите БЕСПЛАТНЫЙ API-ключ).
2. Реализуйте многопоточный сбор информации о вакансиях с сайта hh.ru (если возможно, с применением асинхронных методов). Обязательно сохраняйте информацию об описании вакансии, зарплатной вилке и работодателе.
3. Выполните предварительную обработку текстовых данных об описании вакансии.
4. Сохраните полученный датасет.

## Критерии приема работы

* Сбор датасета производится в нескольких потоках (с использованием асинхронности, если возможно);
* выполнена предварительная обработка данных;
* датасет сохранен в структуру данных/файл (можно прочитать и заргузить датасет в рабочую среду без сбора по API).
